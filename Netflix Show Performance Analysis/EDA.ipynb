```python
import pandas as pd
```

```python
df=pd.read_csv("cleaned_netflix_data.csv")
```

```python
df['type'].value_counts()
```

```python
df['genres'].value_counts().head(10)  # Top 10 most common genres
```

```python
import matplotlib.pyplot as plt

top_genres = df['genres'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_genres.plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Common Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df['releaseYear'].value_counts().sort_index().tail(20)  # Last 20 years
```

```python
df['releaseYear'].value_counts().sort_index().plot(kind='line', figsize=(12,5), color='red')
plt.title("Trend of Netflix Content Releases Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()
```

```python
df['imdbAverageRating'].describe()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['imdbAverageRating'], bins=20, color='purple', edgecolor='black')
plt.title("Distribution of IMDb Ratings")
plt.xlabel("IMDb Rating")
plt.ylabel("Count")
plt.show()
```

```python
df[['title', 'imdbAverageRating']].sort_values(by='imdbAverageRating', ascending=False).head(10)
```

```python
df[['title', 'imdbNumVotes']].sort_values(by='imdbNumVotes', ascending=False).head(10)
```

```python
top_voted = df[['title', 'imdbNumVotes']].sort_values(by='imdbNumVotes', ascending=False).head(10)
plt.figure(figsize=(12,5))
plt.barh(top_voted['title'], top_voted['imdbNumVotes'], color='green')
plt.xlabel("Number of Votes")
plt.ylabel("Show Title")
plt.title("Top 10 Most Voted Shows on IMDb")
plt.gca().invert_yaxis()  # Invert y-axis to show highest at top
plt.show()
```
